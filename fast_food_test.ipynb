```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Name of the CSV file
file ='ETL_final.csv'
```

```python
# define df
food_df ='ETL_final.csv'
```

```python
food_df = pd.read_csv(file)
food_df.head()
```

```python
# Show plot
ax1 = food_df.plot.bar(x="state_name", y=["ff_count"], figsize=(15,4), color='g', align='edge')
ax2 = food_df.plot.line(x="state_name", y="obesity", color='b', rot=90, ax=ax1, secondary_y=True)
plt.xticks(np.arange(len(food_df["state_name"]))+0.2, food_df["state_name"])
plt.xlim(-.5, len(food_df["state_name"]))
ax1.legend(loc='center', bbox_to_anchor=(.06, 1.1), ncol=1)
ax2.legend(loc='center', bbox_to_anchor=(.94, 1.06), ncol=1)
ax1.set_xlabel("")
ax1.set_ylabel('FastFood Location', fontsize=15)
ax2.set_ylabel('Obesity%', fontsize=15)

plt.show()
```
